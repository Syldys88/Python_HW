```python
import sympy
import numpy as np
from sympy.plotting import plot
from sympy import symbols
```

### Построим график заданной функции:

```python
x = symbols('x')
plot(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, (x, -100, 100))
```

### Интервалов возрастания и убывания бесконечное множество.

### Найдем корни:

```python
sympy.solveset(sympy.Eq(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, 0), x)
```

### Промежутков где f > 0 и f < 0 бесконечное множество. 

### Для примера определим промежутки, на котором f > 0 и f < 0 в диапазоне -10 < x < 10 c с шагом 1.

```python
x = np.arange(-10, 10, 1)
func = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

pos_list = []
neg_list = []

for i in x:
    res = -12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30
    if res > 0:
        pos_list.append(res)
    else:
        neg_list.append(res)

print(f'f > 0 в интервале: \n{pos_list[-1]} --> {pos_list[0]}\n')
print(pos_list)
print(f'\nf < 0 в интервале:\n{neg_list[0]} --> {neg_list[-1]}\n')
print(neg_list)
```
